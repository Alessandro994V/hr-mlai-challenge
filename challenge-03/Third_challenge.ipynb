```python
url_position = 'https://nft.17tons.tech/explorer/trees/land/5'
response_position = requests.get(url_position)
data_position = response_position.json()
```

```python
data_position['trees'][0]
```

```python
rows = []

# Loop over each element to flatten and append to rows
for item in data_position['trees']:
    row = {
        '_id': item['_id'],
        'treeId': item['treeId'],
        '__v': item['__v'],
        'description': item['description'],
        'geolocation': item['geolocation'],
        'image': item['image'],
        'landId': item['landId'],
        'lastUpdated': item['lastUpdated'],
        'name': item['name']
    }
    # Flatten attributes
    for attr in item['attributes']:
        row[attr['trait_type']] = attr['value']
    
    rows.append(row)

# Convert to DataFrame
pandas_trees = pd.DataFrame(rows)
```
